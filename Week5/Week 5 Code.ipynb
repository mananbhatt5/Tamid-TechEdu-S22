### Replacing underscores with space bars ###

```python
df_cols = df.columns.str.replace('_', ' ')
df_cols
```

### Making DataFrame smaller ###

```python
df.columns = df_cols
df2 = df[['id', 'ident', 'type', 'name', 'latitude deg', 'longitude deg', 'elevation ft', 'continent', 'iso country','iso region', 'municipality', 'scheduled service', 'local code']]
df2
```

### Reviewing df.loc ###

```python
df3 = df2.loc[df2['scheduled service'].str.contains('yes')]
df3
```

Now we can see that the first element of the df is the 228th column of the og df

```python
df3.index[0]
```

### Making smaller df's using logical operators ###

```python
df4 = df2.loc[df2['name'].str.contains("Airport") & df2['scheduled service'].str.contains("yes")]
df4
```

### n-Dim, Shape, and Size ###

```python
print(df4.ndim) # 2D array
print(df4.name.ndim) # 1D array
print(df4.shape) # rows , cols
print(df4.size) # rows * cols
```

### Drop all NaN's from the dataset ###

```python
df5 = df4.dropna()
df5
```

North America's values all got dropped since NA is also considered a NaN value.

### Making the DF even smaller! ###

```python
df5 = df5.loc[df5['elevation ft'] < 500]
df5
```

### Compare these nDim, Size, and Shape values to the original ###

```python
print(df5.ndim)
print(df5.name.ndim)
print(df5.shape)
print(df5.size)
```
